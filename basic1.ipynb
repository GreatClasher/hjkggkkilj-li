```python
import cv2 as cv
```

```python
cap =cv.VideoCapture(0)
fourcc = cv.VideoWriter_fourcc(*'XVID')
out = cv.VideoWriter('output1112.avi',fourcc,20,(280,280))
while(cap.isOpened()):
    ret,frame = cap.read()
   
    if ret == True:
        out.write(frame)
        cv.imshow('video',frame)
        k = cv.waitKey(1)
        if k == 27:
            break
    else:
        break
   
# img = cv.imread("lena.jpg",cv.IMREAD_COLOR)
# cv.imshow("image",img)
# img2 = cv.line(img,(200,200),(350,350),(250,0,0),2,cv.LINE_AA)
# cv.imshow("image",img2)
# img2 = cv.rectangle(img,(0,0),(250,320),(250,0,0),2,cv.LINE_AA)
# cv.imshow("image",img2)
    
cap.release()   
cv.destroyAllWindows()
    
# elif k == ord('s'):
#     cv.imwrite("lena1111.png",img)
#     cv.destroyAllWindows()
```
